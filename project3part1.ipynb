```python
import pandas as pd
import seaborn as sns
import numpy as np
```

# basics

```python
basics_url = 'https://datasets.imdbws.com/title.basics.tsv.gz'
akas_url = 'https://datasets.imdbws.com/title.akas.tsv.gz'
ratings_url = 'https://datasets.imdbws.com/title.ratings.tsv.gz'

basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
basics.info()
```

```python
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
akas.head()
```

```python
# convert nulls
basics = basics.replace({'\\N':np.nan})
```

```python
# eliminate null runtimeMinutes
basics.dropna(subset=['runtimeMinutes'], inplace=True)
basics.isna().sum()
```

```python
# eliminate nulls in 'genres'

basics.dropna(subset=['genres'], inplace=True)
basics.isna().sum()
```

```python
# keep only titleType == 'movie'
basics['titleType'].value_counts()
movie_filter = basics['titleType'] == 'movie'
basics.loc[movie_filter, :].head()
```

```python
# keep 'startYear' between 2000-2022
yrFilter = basics['startYear'] >= '2000'
yrFilter2 = basics['startYear'] < '2022'
```

```python
basics = basics.loc[movie_filter]
basics = basics.loc[yrFilter & yrFilter2]
basics.info()
basics.head()
```

```python
basics = basics[(basics['genres'].notna())]
```

```python
# eliminate documentaries 
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
basics.info()
basics.head()
```

```python
# save basics df
basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
```

# akas


```python
akas.info()
akas.head()
```

```python
# convert nulls
akas = akas.replace({'\\N':np.nan})
```

```python
us_filter = akas['region'] == 'US'
akas = akas[(us_filter)] 
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
# filter basics
basics = basics[keepers]
basics.head()
```

# Ratings

```python
ratings.info()
ratings.head()
```

```python
# convert nulls
ratings = ratings.replace({'\\N':np.nan})
```

```python
# Filter the ratings table down to only include the US by using the filter akas dataframe
keepers = ratings['tconst'].isin(akas['titleId'])
ratings = ratings[keepers]
ratings.info()
ratings.head()
```

# save DFs to Data folder

```python
## Save all dataframes to file.

akas.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
ratings.to_csv("Data/title_ratings.csv.gz",compression='gzip',index=False)
```
