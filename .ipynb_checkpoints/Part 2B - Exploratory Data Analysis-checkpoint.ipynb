```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
df2 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
```

```python
# concat 2000/2001
df = pd.concat((df1, df2), axis = 0)
```

```python
# filter out movies without revenue or budget > 0 
budg_filter = df['budget'] > 0 
rev_filter = df['revenue'] > 0
df = df.loc[budg_filter | rev_filter]
```

```python
df.info()
df.head()
```

```python
df['certification'].value_counts()
```

## After filtering the data we are left with 728 movies that had valid financial data.

```python
## format y axis for millions
def millions(x,pos):
    new_x = x/1_000_000
    return f"{new_x:,.0f}M"
```

```python
## Make a larger fig/ax before plotting
fig, ax = plt.subplots(figsize=(10,5))

## Plot histogram
sns.histplot(data=df, x='budget', ax=ax)
ax.set(title='Distribution of Movie Budgets');

## Annotating mean and median
mean_budg = df['budget'].mean()
ax.axvline(mean_budg,color='k', ls=':', 
           label=f"Mean Budget= ${round(mean_budg,2)}");

med_budg = df['budget'].median()
ax.axvline(med_budg,color='k', ls=':', 
           label=f"Median budget= ${round(med_budg,2)}");

ax.xaxis.set_major_formatter(millions);

ax.legend();
```

```python
## Make a larger fig/ax before plotting
fig, ax = plt.subplots(figsize=(6,5))

## Plot histogram
ax = sns.histplot(data=df, x='revenue', ax=ax)
ax.set(title='Distribution of Movie Revenues');

## Annotating mean and median
mean_rev = df['revenue'].mean()
ax.axvline(mean_rev,color='k', ls=':', 
           label=f"Mean Revenue= ${round(mean_rev,2)}");

med_rev = df['revenue'].median()
ax.axvline(med_rev,color='k', ls=':', 
           label=f"Median Revenue= ${round(med_rev,2)}");

ax.xaxis.set_major_formatter(millions);


ax.legend();
```

## Number of Movies by Rating

```python
fig = sns.countplot(data=df, x='certification',palette='dark');
plt.title('Number of Movies by MPAA Rating');
```

## Average Budget by Certification

```python
# group avg budget by certification 
certBudget = df.groupby(['certification'])['budget'].mean()
certBudget
```

```python
# plot avg budget by certification 
ax = sns.barplot(x=certBudget.index, y = certBudget.values, palette='muted')
plt.ylabel('Average Budget', fontsize=16)
plt.xlabel('Certification',fontsize=16 )
ax.yaxis.set_major_formatter(millions);
plt.title('Average Budget by Certification', fontsize=18);
plt.savefig('movBudgByCert.jpeg')
```

## Average Revenue by Certification

```python
certRev = df.groupby('certification')['revenue'].mean()
certRev
```

```python
ax = sns.barplot(x=certRev.index, y = certRev.values, palette='muted')
plt.ylabel('Average Revenue', fontsize=16)
plt.xlabel('Certification',fontsize=16 )
ax.yaxis.set_major_formatter(millions);
plt.title('Average Revenue by Certification', fontsize=18);
plt.savefig('movRevByCert.jpeg')
```
